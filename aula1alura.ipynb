```python
# Create a dictionary with the current column names and their Portuguese translations
new_column_names = {
    'ano': 'ano',
    'nivel_experiencia': 'nivel_experiencia',
    'tipo_emprego': 'tipo_emprego',
    'titulo_cargo': 'cargo',
    'salario': 'salario',
    'moeda_salario': 'moeda',
    'salario_em_us': 'usd',
    'residencia_empregado': 'residencia',
    'proporcao_remoto': 'remoto',
    'localizacao_empresa': 'localizacao_empresa',
    'porte_empresa': 'porte_empresa'
}

# Rename the columns using the dictionary
df.rename(columns=new_column_names, inplace=True)

# Display the updated column names to confirm the changes
display(df.columns)
```

```python
# Create a dictionary to map the English experience levels to Portuguese
experience_level_map = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Jnior',
    'EX': 'Executivo'
}

# Apply the mapping to the 'nivel_experiencia' column
df['nivel_experiencia'] = df['nivel_experiencia'].replace(experience_level_map)

# Display the updated value counts to confirm the changes
display(df['nivel_experiencia'].value_counts())
```

```python
# Create a dictionary to map the English employment types to Portuguese
employment_type_map = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

# Apply the mapping to the 'tipo_emprego' column using replace
df['tipo_emprego'] = df['tipo_emprego'].replace(employment_type_map)

# Display the updated value counts to confirm the changes
display(df['tipo_emprego'].value_counts())
```

```python
# Create a dictionary to map the remote ratio to Portuguese terms
remote_map = {
    0: 'Presencial',
    50: 'Hbrido',
    100: 'Remoto'
}

# Apply the mapping to the 'remoto' column using replace
df['remoto'] = df['remoto'].replace(remote_map)

# Display the updated value counts to confirm the changes
display(df['remoto'].value_counts())
```
